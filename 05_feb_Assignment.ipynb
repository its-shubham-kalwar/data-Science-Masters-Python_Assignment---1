```python
# Question - 1
# Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
```

```python
In Object-Oriented Programming (OOP), a class is a blueprint or a template for creating objects. 
It defines a set of attributes and behaviors that an object of that class should have. 
An object, on the other hand, is an instance of a class. 
It is a concrete representation of the class that has been created in memory and can be 
manipulated using its attributes and methods.

Let's take an example to understand this better. Consider the class "Car". 
It can have attributes such as "color", "model", "year", "price", etc. 
It can also have behaviors such as "start()", "stop()", "accelerate()", and "brake()". 
These attributes and behaviors are defined in the class.

Now, let's create an object of the class "Car". We can create an object named "myCar" 
with the following code:


class Car:
    def __init__(self, color, model, year, price):
        self.color = color
        self.model = model
        self.year = year
        self.price = price
    
    def start(self):
        print("Starting the car...")
        

myCar = Car("blue", "Toyota", 2021, 25000)

In the code above, we defined the class "Car" with the attributes "color", "model", "year", and 
"price" and the behavior "start()". We then created an object named "myCar" of the class "Car" 
and assigned it the values "blue", "Toyota", 2021, and 25000 for its attributes.

We can now use the methods of the object "myCar" to manipulate its attributes. 
For example, we can start the car by calling the "start()" method of the object:


myCar.start()

This will output the following:


Starting the car...
In summary, a class is a blueprint that defines the attributes and behaviors that an object of
that class should have, while an object is an instance of a class that has been created in memory 
and can be manipulated using its attributes and methods.
```

```python
# Question - 2 
# Name the four pillars of OOPs.
```

```python
Encapsulation: Encapsulation is the mechanism that binds the data and the methods that manipulate 
the data into a single unit. It ensures that the internal representation of an object is hidden 
from the outside world and can only be accessed through a public interface.

Inheritance: Inheritance allows a class to inherit properties and methods from a parent class. 
It enables code reuse and promotes the creation of a hierarchical class structure.

Polymorphism: Polymorphism is the ability of an object to take on many forms. 
It allows objects of different classes to be treated as if they were of the same class, 
thus making the code more flexible and extensible.

Abstraction: Abstraction is the process of hiding complex implementation details and presenting 
only the essential features of an object to the outside world. It enables the creation of simple 
and easy-to-use interfaces for complex systems.
```

```python
# Question - 4
# Why self is used in OOPs?
```

```python
In Object-Oriented Programming (OOP), self is a reference to the instance of the class 
that is currently being manipulated. It is a parameter that is passed to instance methods 
and is used to access and modify the object's attributes.

When an object of a class is created, the object is stored in memory, and a reference to that 
object is returned. This reference is passed to the self parameter in all instance methods defined 
in that class. The self parameter is used to access and modify the object's attributes.

For example, consider the following class definition:


class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def say_hello(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")
        
        
In the code above, we defined a class named "Person" with two attributes: "name" and "age". 
We also defined an instance method named "say_hello" that uses the self parameter to access 
the object's attributes and print a greeting message.

When we create an object of the class "Person" and call the say_hello() method, the self parameter 
is automatically passed to the method, and it can access and modify the object's attributes:


person = Person("John", 30)
person.say_hello()
This will output the following:


Hello, my name is John and I am 30 years old.
In summary, self is used in OOP to reference the instance of the class that is currently being manipulated.
It is passed as a parameter to instance methods and is used to access and modify the object's attributes.
```

```python
Inheritance is a mechanism in Object-Oriented Programming (OOP) that allows a class to inherit properties 
and methods from a parent class. It enables code reuse and promotes the creation of a hierarchical 
class structure.

There are several types of inheritance in OOP:

Single inheritance: In single inheritance, a class inherits properties and methods from a single parent 
class.

Example:

class Animal:
    def __init__(self, name):
        self.name = name
        
    def eat(self):
        print("Eating...")

class Dog(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def bark(self):
        print("Woof!")

dog = Dog("Fido")
dog.eat()
dog.bark()

In the code above, we defined a class named "Animal" with a constructor method and an instance method
named "eat". We also defined a class named "Dog" that inherits from the "Animal" class using the Animal 
keyword in the class definition. The Dog class has a constructor method that calls the constructor method 
of the parent class using the super() function. The Dog class also has an instance method named "bark". 
When we create an object of the Dog class and call its methods, it can access the methods and attributes
of the parent class.


Multi-level inheritance: In multi-level inheritance, a class inherits from a parent class, which in turn 
inherits from another parent class.

Example:

class Animal:
    def __init__(self, name):
        self.name = name
        
    def eat(self):
        print("Eating...")

class Mammal(Animal):
    def __init__(self, name):
        super().__init__(name)
        
    def walk(self):
        print("Walking...")

class Dog(Mammal):
    def __init__(self, name):
        super().__init__(name)
        
    def bark(self):
        print("Woof!")

dog = Dog("Fido")
dog.eat()
dog.walk()
dog.bark()

In the code above, we defined a class named "Animal" with a constructor method and an instance method 
named "eat". We also defined a class named "Mammal" that inherits from the "Animal" class and has an 
instance method named "walk". Finally, we defined a class named "Dog" that inherits from the "Mammal" 
class and has an instance method named "bark". When we create an object of the Dog class and call its 
methods, it can access the methods and attributes of both the Animal and Mammal classes.

Multiple inheritance: In multiple inheritance, a class inherits properties and methods from multiple 
parent classes.

Example:

class Animal:
    def __init__(self, name):
        self.name = name
        
    def eat(self):
        print("Eating...")

class Bird:
    def fly(self):
        print("Flying...")

class Parrot(Animal, Bird):
    def __init__(self, name):
        super().__init__(name)
        
    def talk(self):
        print("Polly wants a cracker!")

parrot = Parrot("Polly")
parrot.eat()
parrot.fly()
parrot.talk()

In the code above, we defined a class named "Animal" with a constructor method and an instance method 
named "eat". We also defined a class named "Bird" with an instance method named "fly". 
Finally, we defined a class named "Parrot" that inherits from both the "Animal" and "Bird" classes 
and has an instance method named "talk". When we create an object of the `
```
